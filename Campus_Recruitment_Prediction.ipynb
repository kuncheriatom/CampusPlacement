# Data Preprocessing


```python
df.isna().sum()
```

```python
not_placed_count = df['status'].value_counts()['Not Placed']
print(not_placed_count)
```

Dropping the salary Field as the salary is only for the placed students

```python
df = df.drop(columns=['salary'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split



# Split the data into training and test sets
X = df.drop(columns=['status'])
y = df['status']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,StandardScaler
from sklearn.compose import ColumnTransformer

# Define categorical and numerical features
categorical_features = ['gender', 'ssc_b', 'hsc_b', 'hsc_s', 'degree_t', 'workex', 'specialisation']
numerical_features = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']

# Preprocessor with both one-hot encoding for categorical features and standard scaling for numerical features
column_transformer = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first'), categorical_features),
        ('num', StandardScaler(), numerical_features)
    ])

# Fit and transform the training data
X_train_encoded = column_transformer.fit_transform(X_train)

# Transform the test data
X_test_encoded = column_transformer.transform(X_test)

# Encode the target variable
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

print("Encoded Training data shape:", X_train_encoded.shape)
print("Encoded Test data shape:", X_test_encoded.shape)



```

## Logistic Regression Model and Hyperparameter Tuning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Initialize the model
log_clf = LogisticRegression(max_iter=1000, random_state=42)

# Define hyperparameters for tuning
log_params = {'C': [0.1, 1, 10]}

# Perform grid search
log_grid = GridSearchCV(log_clf, log_params, cv=5, scoring='accuracy')

# Fit the model
log_grid.fit(X_train_encoded, y_train_encoded)

# Best estimator
log_best = log_grid.best_estimator_

print("Best Logistic Regression parameters:", log_grid.best_params_)
```

## Random Forest Model and Hyperparameter Tuning

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
rf_clf = RandomForestClassifier(random_state=42)

# Define hyperparameters for tuning
rf_params = {'n_estimators': [50, 100, 200], 'max_depth': [2, 10, 20]}

# Perform grid search
rf_grid = GridSearchCV(rf_clf, rf_params, cv=5, scoring='accuracy')

# Fit the model
rf_grid.fit(X_train_encoded, y_train_encoded)

# Best estimator
rf_best = rf_grid.best_estimator_

print("Best Random Forest parameters:", rf_grid.best_params_)
```

# Gradient Boosting Model and Hyperparameter Tuning

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize the model
gb_clf = GradientBoostingClassifier(random_state=42)

# Define hyperparameters for tuning
gb_params = {'n_estimators': [50, 100, 200], 'learning_rate': [0.01, 0.1, 0.2]}

# Perform grid search
gb_grid = GridSearchCV(gb_clf, gb_params, cv=5, scoring='accuracy')

# Fit the model
gb_grid.fit(X_train_encoded, y_train_encoded)

# Best estimator
gb_best = gb_grid.best_estimator_

print("Best Gradient Boosting parameters:", gb_grid.best_params_)
```

## Support Vector Machine Model and Hyperparameter Tuning

```python
from sklearn.svm import SVC

# Initialize the model
svc_clf = SVC(random_state=42)

# Define hyperparameters for tuning
svc_params = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}

# Perform grid search
svc_grid = GridSearchCV(svc_clf, svc_params, cv=5, scoring='accuracy')

# Fit the model
svc_grid.fit(X_train_encoded, y_train_encoded)

# Best estimator
svc_best = svc_grid.best_estimator_

print("Best SVC parameters:", svc_grid.best_params_)
```

## k-Nearest Neighbour  Model and Hyperparameter Tuning

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the model
knn_clf = KNeighborsClassifier()

# Define hyperparameters for tuning
knn_params = {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}

# Perform grid search
knn_grid = GridSearchCV(knn_clf, knn_params, cv=5, scoring='accuracy')

# Fit the model
knn_grid.fit(X_train_encoded, y_train_encoded)

# Best estimator
knn_best = knn_grid.best_estimator_

print("Best k-NN parameters:", knn_grid.best_params_)
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Function to evaluate model
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', 
                xticklabels=['Not Placed', 'Placed'], yticklabels=['Not Placed', 'Placed'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
    
    print(f'Accuracy: {accuracy:.2f}')
    print(f'Precision: {precision:.2f}')
    print(f'Recall: {recall:.2f}')
    print(f'F1-score: {f1:.2f}')
   
   


```

```python

print("Logistic Regression:")
evaluate_model(log_best, X_test_encoded, y_test_encoded)
```

```python
print("\nRandom Forest:")
evaluate_model(rf_best, X_test_encoded, y_test_encoded)
```

```python
print("\nGradient Boosting:")
evaluate_model(gb_best, X_test_encoded, y_test_encoded)
```

```python
print("\nSVC:")
evaluate_model(svc_best, X_test_encoded, y_test_encoded)
```

```python
print("\nk-NN:")
evaluate_model(knn_best, X_test_encoded, y_test_encoded)
```

# Voting Classifier

```python
from sklearn.ensemble import VotingClassifier

# Voting Classifier
voting_clf = VotingClassifier(estimators=[
    ('lr', log_best),
    ('rf', rf_best),
    ('gb', gb_best),
    ('svc', svc_best),
    ('knn', knn_best)
], voting='hard')

# Train the voting classifier
voting_clf.fit(X_train_encoded, y_train_encoded)

print("\nVoting Classifier:")
evaluate_model(voting_clf, X_test_encoded, y_test_encoded)
```

# comparison of model performances

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Define a function to evaluate models
def evaluate_model(model, X_test, y_test):
    
    # Predict labels
    y_pred = model.predict(X_test)
    
    # Calculate evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    # Generate confusion matrix
    cm = confusion_matrix(y_test, y_pred)
   
    
    # Return metrics as dictionary
    return {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-score': f1
    }
```

```python
# Dictionary of models to evaluate
models = {
    'Logistic Regression': log_best,
    'Random Forest': rf_best,
    'Gradient Boosting': gb_best,
    'Support Vector Machine': svc_best,
    'k-Nearest Neighbors': knn_best,
    'Voting Classifier': voting_clf
}


# Initialize an empty dictionary to store results
results = {}

# Evaluate each model and store results
for name, model in models.items():
    results[name] = evaluate_model(model, X_test_encoded, y_test_encoded)


# Convert results dictionary to DataFrame
metrics_df = pd.DataFrame(results).T

print(metrics_df)

```

```python

# Plotting comparison of model performances for multiple metrics
plt.figure(figsize=(14, 10))

# Plot accuracy
plt.subplot(2, 2, 1)
sns.barplot(x=metrics_df.index, y='Accuracy', data=metrics_df, palette='viridis') # type: ignore
plt.xticks(rotation=45)
plt.title('Model Performance Comparison - Accuracy')
plt.ylabel('Accuracy')

# Plot precision
plt.subplot(2, 2, 2)
sns.barplot(x=metrics_df.index, y='Precision', data=metrics_df, palette='viridis')
plt.xticks(rotation=45)
plt.title('Model Performance Comparison - Precision')
plt.ylabel('Precision')

# Plot recall
plt.subplot(2, 2, 3)
sns.barplot(x=metrics_df.index, y='Recall', data=metrics_df, palette='viridis')
plt.xticks(rotation=45)
plt.title('Model Performance Comparison - Recall')
plt.ylabel('Recall')

# Plot F1-score
plt.subplot(2, 2, 4)
sns.barplot(x=metrics_df.index, y='F1-score', data=metrics_df, palette='viridis')
plt.xticks(rotation=45)
plt.title('Model Performance Comparison - F1-score')
plt.ylabel('F1-score')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

The Gradient Boost is performing the best among all the models.
